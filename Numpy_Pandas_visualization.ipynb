```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snr
```

Assignment 1


```python
data=pd.DataFrame(np.random.rand(5,6),index=['r1','r2','r3','r4','r5'],columns=['c1','c2','c3','c4','c5','c6'])
```

```python
print(data)
```

```python
#bar graph
```

```python
data.plot(kind='bar')
```

```python
#pie chart
```

```python
data.iloc[0:5].plot.pie(figsize=(15,15),subplots=True,fontsize=12,autopct='%.2f')
```

```python
#histogram
```

```python
data['c5'].diff().plot(kind='hist',bins=200)
```

```python
#line graph
```

```python
line=fitbit.plot(figsize=(10,5),title='fitbit data')
line.set_xlabel('X axis')
line.set_ylabel('Y axis')
```

```python
snr.relplot(x='TotalSteps',y='TotalDistance',data=fitbit,hue='VeryActiveDistance')
```

```python
snr.relplot(x='TotalSteps',y='TotalDistance',data=fitbit,col='Calories')
```
