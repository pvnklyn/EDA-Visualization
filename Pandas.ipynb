Convert these lists into a multinested array and fetch the given rows and columns

1. arr=[0:2,2:]
2. arr=[0:2,0:2]
3. arr=[1:,:4]

```python
import numpy as np
```

```python
my_list1=[4,5,6,7,8]

my_list2=[-2,6,8,3,4]

my_list3=[5,6,0,-3,5]
```

```python
arr=np.array([my_list1,my_list2,my_list3])
```

```python
print(arr)
```

```python
print(arr[0:2,2:])
```

```python
print(arr[0:2,0:2])
```

```python
print(arr[1:,:4])
```
